En pulsos analizados en `pulAnl160211` y en `pulAnl160107` erré la corriente de entrada para comparar con la calibración de `anl150807`. 

Busco levantar curvas de $\Delta t$ vs. $\tau$ usando generadores de pulsos cuadrados varios:
    - BL-2
    - HP 3310B
    
Los pulsos cuadrados $\Delta t$ barriran desde $50\,\mathrm{ns}$ a $250\,\mathrm{ns}$.
Los recorrro en forma descendente.

Las corridas se realizan con el código `pulseTek2002B`

- Function = (pulso negativo)
