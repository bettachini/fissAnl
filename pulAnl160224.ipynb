El error relativo de $\tau$ es mayor que el de $\Delta t$. Tendría que hacer el ajuste con el primero como variable dependiente. 

Por ahora sigo con el procedimiento anterior y dejo esta inquietud para después.

El peso del error como 
$w_i= 1/\sigma_i^2$
es aquel que logra que los coeficientes del ajuste lineal sea los mejores estimadores no sesgados(best linear unbiased estimator) `https://en.wikipedia.org/wiki/Least_squares#Weighted_least_squares`

```python
w_DB2_anl150807= 1./(DB2_anl150807['ePulser'])**2
z3_DB2_anl150807, mCov3_DB2_anl150807 = np.polyfit(DB2_anl150807['tau'], DB2_anl150807['pulser'], 3, w= w_DB2_anl150807, cov=True)
p3_DB2_anl150807 = np.poly1d(z3_DB2_anl150807)
errZ3_DB2_anl150807= np.sqrt(np.diag(mCov3_DB2_anl150807))
```

```python
w_DB2_pulAnl160107= 1./ DB2_pulAnl160107['anchoPulserErrorS']**2
z3_DB2_pulAnl160107, mCov3_DB2_pulAnl160107 = np.polyfit( DB2_pulAnl160107['tauAMCS'], DB2_pulAnl160107['anchoPulserS'], 3, w= w_DB2_pulAnl160107, cov=True)
p3_DB2_pulAnl160107 = np.poly1d( z3_DB2_pulAnl160107)
errZ3__DB2_pulAnl160107= np.sqrt( np.diag( mCov3_DB2_pulAnl160107 ) )
```

```python
w_HP3310B_pulAnl160224= 1./ HP3310B_pulAnl160224['anchoPulserError3310B']**2
z3_HP3310B_pulAnl160224, mCov3_HP3310B_pulAnl160224 = np.polyfit( HP3310B_pulAnl160224['tauAMC3310B'], HP3310B_pulAnl160224['anchoPulser3310B'], 3, w= w_HP3310B_pulAnl160224, cov=True)
p3_HP3310B_pulAnl160224 = np.poly1d( z3_HP3310B_pulAnl160224)
errZ3__HP3310B_pulAnl160224= np.sqrt( np.diag( mCov3_HP3310B_pulAnl160224 ) )
```
