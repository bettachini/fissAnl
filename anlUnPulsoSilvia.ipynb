```python
# Rutinas de graficación mínimas

def plotaSerie(conv1, T, Meas2, messungenZahl, prop, bMeas2):
    # Reemplazo de plota3
    '''
    Grafica en valores físicos de ambos canales, niveles de base (izquierda), máximos, 10% y 90% del risetime

    (T, Meas2) (maxCH1_index, maxCH2_index, iProcent10, iProcent90) -> 
    '''
    # Gráfica max, nivel cero, etc.
    fig, (ax0, ax1) = plt.subplots(nrows=2, sharex=True)
    fig.set_size_inches(10,8)
    fig.subplots_adjust(hspace=0.000)
    # Axis 0
    ax0.plot(T, (4E-2* Meas2[0,:]+ conv1.ceroCh1Div)* ((-1)* conv1.escalaCh1VoltsDiv), 'y')  # puesto que ch1= ch1* ()
    # Make the y-axis label and tick labels match the line color.
    ax0.set_ylabel('ch 1 [V]', color='y')
    ax0.grid(True)
    # Axis 1
    ax1.plot(T, (4E-2* Meas2[1,:]- conv1.ceroCh2Div)* conv1.escalaCh2VoltsDiv, 'c')
    ax1.set_ylabel('ch 2 [V]', color='c')
    ax1.set_xlabel('tiempo [s]')
    ax1.grid(True)
    ## zero level lines
    erstePunkten= int(messungenZahl[1]* prop) # auxiliar
    zeroT= T[0:(erstePunkten- 1)]
    zeroAx0= np.ones([erstePunkten- 1, 1])* (4E-2* bMeas2[0]+ conv1.ceroCh1Div)* ((-1)* conv1.escalaCh1VoltsDiv)
    zeroAx1= np.ones([erstePunkten- 1, 1])* (4E-2* bMeas2[1]- conv1.ceroCh2Div)* conv1.escalaCh2VoltsDiv
    ax0.plot(zeroT, zeroAx0, 'k-')
    ax1.plot(zeroT, zeroAx1, 'k-')
    #plt.savefig('1090prozent.pdf')
    plt.show()
    

def tauCetiSerie(tipo, T, Meas2, messungenZahl, prop, conv1):
    # Reemplazo de tauCeti4
    '''
    Calcula índices risetime como tiempo entre 10% y 90% de excursión entre nivel de base y máximo.
    Retorna area por sobre nivel de base de AMC entre 10% a izquierda y derecha.
   
    Meas2: array (2,:) con Meas2[0] canal 1, Meas2[0] canal 2, producto de np.vstack
    baseMeas2: base(Meas2)
    '''
    bMeas2= baseMeas2(Meas2, messungenZahl, prop) # niveles base
    plotaSerie(conv1, T, Meas2, messungenZahl, prop, bMeas2)
    

def tauAdqSerie(tipo, npzData, prop):
    # reemplazo de tausAreasMaxSerieAdq
    '''
    Vectores risetime canal AMC de archivo acq comprimido
    
    npzData= numpy.lib.npyio.NpzFile
    prop= porcentaje inferior del canal a promediar para obtener nivel de base     
    '''
    FLUC= npzData['ch1']
    AMC= npzData['ch2']
    messungenZahl= AMC.shape
    # tiempo
    T= npzData['zeit']#
    conv1= parametrosConversion(npzData)
    i=0
    q=0
    cucho= (messungenZahl[0]- q )
    # cucho= (messungenZahl[0]- q- 1 )
    while (i< cucho ): # recorre cada forma de onda en medición npzData
        # print('Adqusición actual= ', int(i))
        Meas2= np.array([FLUC[i+q], AMC[i+q] ] )
        # anchoPulseri, areaPulseri, maxPulseri, tauAMCi, areaTotalAMCi, areaSubidaAMCi, maxAMCi= tauCeti4(tipo, T, Meas2, messungenZahl, prop, conv1, rechazo)
        tauCetiSerie(tipo, T, Meas2, messungenZahl, prop, conv1)
        i+=1
        cucho= (messungenZahl[0]- q )
        
        
def agregadorSerie(tipo, anne, mes, dia, rango):
    # reemplazo de agregador2
    for i in rango:
        acq= np.load(dataPathFile2(tipo, anne, mes, dia, i))
        # anchoPulser, areaPulser, maxPulser, tauAMC, areaSubidaAMC, areaTotalAMC, maxAMC= tausAreasMaxSerieAdq(tipo, acq, prop, rechazo)
        tauAdqSerie(tipo, acq, prop)
```
